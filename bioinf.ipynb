Series GSE57957: Expression profile of Hepatocellular Carcinoma

```python
import pandas as pd
```

```python
with open ("GSE57956.top.table.tsv") as GSE57956:
    meth = GSE57956.readlines() 
    
meth_liste = []

for m in meth:
    meth_liste.append(m.split('\t'))
    
methylation = pd.DataFrame(meth_liste[1:], columns=meth_liste[0])
```

```python
with open ("GSE57957.top.table.tsv") as GSE57957:
    exp = GSE57957.readlines() 
    
exp_liste = []

for e in exp:
    exp_liste.append(e.split('\t'))
    
expression = pd.DataFrame(exp_liste[1:], columns=exp_liste[0])
```

```python
for i in range(0, len(methylation["adj.P.Val"])):
    if float(methylation["adj.P.Val"][i]) > 0.05:
        methylation =methylation.loc[:i]
        break
        
for i in range(0, len(expression["adj.P.Val"])):
    if float(expression["adj.P.Val"][i]) > 0.05:
        expression =expression.loc[:i]
        break
```

```python
methylation_down = methylation.copy()
methylation_up = methylation.copy()

for i in range(0, len(methylation_down["adj.P.Val"])):
    if float(methylation_down["logFC"][i]) > -0.2:
        methylation_down = methylation_down.drop([i])        
        
for i in range(0, len(methylation_up["adj.P.Val"])):
    if float(methylation_up["logFC"][i]) < 0.2:
        methylation_up = methylation_up.drop([i])  
```

```python
expression_down = expression.copy()
expression_up = expression.copy()

for i in range(0, len(expression_down["adj.P.Val"])):
    if float(expression_down["logFC"][i]) > 0:
        expression_down = expression_down.drop([i])        
        
for i in range(0, len(expression_up["adj.P.Val"])):
    if float(expression_up["logFC"][i]) < 0:
        expression_up = expression_up.drop([i])  
```
